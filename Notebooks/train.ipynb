```python
import os
import cv2
import numpy as np


text_file_path = '/content/drive/MyDrive/Potholes/yolov7/runs/detect/exp8/labels'
img_file_path = '/content/drive/MyDrive/Potholes/dataset/test_images'


for label in os.listdir(text_file_path):
    label_name = label.split('.')[0]
    print(label)
    ImageID = f'{label_name}.jpg'
    img = cv2.imread(os.path.join(img_file_path,ImageID))
    i_h = img.shape[0]
    i_w = img.shape[1]

    with open(os.path.join(text_file_path,label), 'r') as f:
        lines = f.readlines()
        for line in lines:
            line = np.float64(line.split(" "))
            cx = line[1] * i_w
            cy = line[2] * i_h
            w  = line[3] * i_w 
            h  = line[4] * i_h
            Conf = line[5]
            XMin = max((cx - w/2), 0)
            YMin = max((cy - h/2), 0)
            XMax = max((cx + w/2), 0)
            YMax = max((cy + h/2), 0)

            if os.path.exists('test_labels.csv'):
                with open('test_labels.csv', 'a') as f:
                    text = "\n" + str(XMin) + ',' +  str(YMin) + ',' +  str(XMax) + ',' +  str(YMax) + ',' +  str(Conf) + ',' + str('pothole') + ',' + str(ImageID)
                    f.write(text)
            else:
                with open('test_labels.csv', 'w') as f:
                    text =  'XMin' + ',' +  'YMin' + ',' +  'XMax' + ',' +  'YMax' + ',' +  'Conf' + ',' + 'LabelName' + ',' + 'ImageID'
                    f.write(text)
                    
```
